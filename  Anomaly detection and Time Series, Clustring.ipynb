1Q: What is Dimensionality Reduction? Why is it important in machine
learning?

Ans: Dimensionality reduction is the process of reducing the number of input variables or features in a dataset while retaining the most important information.
It transforms high-dimensional data into a lower-dimensional form without losing essential patterns or relationships.

Importance in machine learning:

-Reduces computational cost: Fewer features mean faster training and prediction.

-Reduces overfitting: Removes redundant or irrelevant features, improving model generalization.

-Improves model performance: Helps algorithms work better on complex and noisy datasets.

-Enhances visualization: Makes it possible to visualize high-dimensional data in 2D or 3D.

2Q:  Name and briefly describe three common dimensionality reduction techniques.

Ans:

(i)Principal Component Analysis (PCA):
A statistical technique that converts correlated features into a smaller set of uncorrelated components called principal components.

(ii)t-Distributed Stochastic Neighbor Embedding (t-SNE):
A nonlinear method mainly used for visualization; preserves local similarities and creates meaningful 2D/3D embeddings.

(iii)Linear Discriminant Analysis (LDA):
A supervised method that reduces dimensions by maximizing the separation between different classes.

3Q: What is clustering in unsupervised learning? Mention three popular clustering algorithms.

Ans: Clustering is an unsupervised learning technique that groups data points into clusters based on similarity. Points within the same cluster are more similar to each other than to points in other clusters.

Popular clustering algorithms:

-K-Means Clustering

-Hierarchical Clustering

-DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

4Q: Explain the concept of anomaly detection and its significance.

Ans:Anomaly detection refers to identifying items, events, or observations that deviate significantly from the normal pattern or expected behavior.

Significance:

-Helps detect fraud, network intrusions, and security breaches.

-Identifies manufacturing defects or equipment failures.
Detects unusual patterns in healthcare, finance, and IoT sensor systems.

-Improves reliability, safety, and decision-making by flagging abnormal events.

5Q:  List and briefly describe three types of anomaly detection techniques.

Ans:

(i)Statistical Methods:

Detect anomalies using probability distributions, thresholds, or statistical models like mean, variance, Z-scores, or confidence intervals.

(ii)Machine Learning Models:

Use algorithms like Isolation Forest, One-Class SVM, or Autoencoders to identify abnormal data points.

(iii)Distance/Clustering-Based Techniques:

Anomalies are recognized as points far from cluster centers or regions of high density (e.g., using K-Means, DBSCAN).

6Q: What is time series analysis? Mention two key components of time series data.

Ans: Time series analysis involves studying data points collected or recorded at regular time intervals to identify trends, patterns, and forecasting future values.

Key components:

-Trend: Long-term upward or downward movement in the data.

-Seasonality: Repeating patterns or cycles occurring at fixed intervals (e.g., daily, monthly).

7Q: Describe the difference between seasonality and cyclic behavior in time series.

Ans:

Seasonality:

-Occurs at fixed, known intervals (daily, weekly, monthly). It is predictable and repeats regularly.

Cyclic Behavior:

-Occurs at irregular intervals and depends on economic or natural cycles. Cycle length is not fixed and is harder to predict.



```python
#8Q: Write Python code to perform K-means clustering on a sample dataset.


# K-Means Clustering Example

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Generate sample data
X, y = make_blobs(n_samples=300, centers=3, random_state=42)

# Apply K-Means
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

# Predicted cluster labels
labels = kmeans.predict(X)

# Plot the clusters
plt.scatter(X[:,0], X[:,1], c=labels)
plt.scatter(kmeans.cluster_centers_[:,0],
            kmeans.cluster_centers_[:,1],
            marker="X", s=200)
plt.title("K-Means Clustering")
plt.show()
```

```python
#9Q: What is inheritance in OOP? Provide a simple example in Python.

#Ans: Inheritance is an Object-Oriented Programming (OOP) concept where one class (child class) acquires the properties and methods of another class (parent class).
#It promotes code reusability and logical organization.

#Example:

class Animal:
    def speak(self):
        return "This animal makes a sound"

class Dog(Animal):
    def speak(self):
        return "The dog barks"

# Usage
d = Dog()
print(d.speak())
```

10Q: How can time series analysis be used for anomaly detection?

Ans: Time series analysis detects anomalies by identifying points that do not fit the expected trend, seasonality, or pattern.

Ways it is used:

-Forecasting-based detection: Compare actual values with predicted values; large deviations indicate anomalies.

-Seasonal decomposition: Identify abnormal peaks or drops after removing trend and seasonal components.

-Statistical thresholds: Mark data points outside confidence intervals as anomalies.

-Model-based approaches: LSTM, ARIMA, and Autoencoders learn temporal patterns and detect irregularities.
